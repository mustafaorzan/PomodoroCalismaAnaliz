```python
import pandas as pd 
import numpy as np

df = pd.read_excel("Çalışma Çizelgesi v2.xlsx")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df['Tarih'] = pd.to_datetime(df['Yıl'].astype(str) + '-' + df['Ay'].astype(str) + '-' + df['Gün'].astype(str))
df['Tarih'] = df['Tarih'].dt.strftime('%d-%m-%Y')
df.drop(columns=['Gün', 'Ay','Yıl'])
df = df[['Tarih', 'GünAdı', 'Yapılan', 'Hedef', 'ÇalışmaOranı', 'FreeDay', 'OffDay']]
df.head()
```

```python
df.tail()
```

```python
df.dropna()
```

```python
df['Yapılan'] = pd.to_numeric(df['Yapılan'], errors='coerce').astype('Int64')
```

```python
df['Yapılan'].sum() / df['Hedef'].sum()
```

```python
toplam_calisma_gunluk_toplam = df.groupby('GünAdı')[['Yapılan','Hedef']].sum().reindex(['Pazartesi', 'Salı', 'Çarşamba', 'Perşembe', 'Cuma', 'Cumartesi', 'Pazar']).reset_index()
toplam_calisma_gunluk_toplam['Çalışma %'] = np.round( ((toplam_calisma_gunluk_toplam['Yapılan'] / toplam_calisma_gunluk_toplam['Hedef']) * 100) ,1) 
toplam_calisma_gunluk_toplam
```

```python
toplam_calisma_gunluk_ortalama = df.groupby('GünAdı')[['Yapılan','Hedef']].mean().reindex(['Pazartesi', 'Salı', 'Çarşamba', 'Perşembe', 'Cuma', 'Cumartesi', 'Pazar']).reset_index()
toplam_calisma_gunluk_ortalama[['Yapılan', 'Hedef']] = np.round(toplam_calisma_gunluk_ortalama[['Yapılan', 'Hedef']], 1)
toplam_calisma_gunluk_ortalama
```

```python
toplam_calisma_gunluk_ortalama[['Yapılan', 'Hedef']].sum()
```

```python
gunluk_bos = df.groupby('GünAdı')[['FreeDay','OffDay']].sum().reindex(['Pazartesi', 'Salı', 'Çarşamba', 'Perşembe', 'Cuma', 'Cumartesi', 'Pazar']).reset_index()
gunluk_bos
```

```python
toplamBos = gunluk_bos['FreeDay'].sum() + gunluk_bos['OffDay'].sum()
toplamBos
```

```python
df.shape[0]-1 #Son gün boş kayıt olduğu için -1 yaptım
```

```python
np.round((toplamBos / (df.shape[0]-1)),2) #Boş Geçen gün oranı %37 (30.Kasım.2023 e kadar)
```

```python
gunluk_bos_ort = df.groupby('GünAdı')[['FreeDay','OffDay']].mean().reindex(['Pazartesi', 'Salı', 'Çarşamba', 'Perşembe', 'Cuma', 'Cumartesi', 'Pazar']).reset_index()
gunluk_bos_ort[['FreeDay', 'OffDay']] = np.round(gunluk_bos_ort[['FreeDay', 'OffDay']], 1)
gunluk_bos_ort
```

```python
gunluk_rapor = toplam_calisma_gunluk_toplam[['GünAdı','Yapılan', 'Hedef','Çalışma %']]
gunluk_rapor[['YapılanOrt', 'HedefOrt']] = toplam_calisma_gunluk_ortalama[['Yapılan', 'Hedef']]
gunluk_rapor = gunluk_rapor[['GünAdı','Yapılan', 'YapılanOrt', 'Hedef', 'HedefOrt', 'Çalışma %']]
gunluk_rapor['FreeDay'] = gunluk_bos['FreeDay']
gunluk_rapor['FreeDayOrt'] = gunluk_bos_ort['FreeDay']
gunluk_rapor['OffDay'] = gunluk_bos['OffDay']
gunluk_rapor['OffDayOrt'] = gunluk_bos_ort['OffDay']
gunluk_rapor
```

```python
sifir_verim = df.loc[ (df['Yapılan'] == 0) & (df['Hedef']>0) ]
sifir_verim.head(3)
```

```python
sifir_verim.shape
```

```python
sifirVerim = sifir_verim.groupby('GünAdı')['Yapılan'].count().reindex(['Pazartesi', 'Salı', 'Çarşamba', 'Perşembe', 'Cuma', 'Cumartesi', 'Pazar']).reset_index()
sifirVerim
```

```python
sifir_verim[sifir_verim['GünAdı']=='Cuma']
```

```python
gunluk_rapor['Yapılmayan'] = sifirVerim['Yapılan']
gunluk_rapor['ToplamBOS'] = gunluk_rapor['Yapılmayan'] + gunluk_rapor['FreeDay'] + gunluk_rapor['OffDay']
gunluk_rapor
```

```python
df['Tarih'] = pd.to_datetime(df['Tarih'], format='%d-%m-%Y')
grouped_df = df.groupby(df['Tarih'].dt.to_period("M")).agg({'Yapılan': 'sum', 'Hedef': 'sum', 'FreeDay': 'sum', 'OffDay': 'sum'}).reset_index()
grouped_df['ÇalışmaOranı'] = grouped_df['Yapılan'] / grouped_df['Hedef'] 
grouped_df['ÇalışmaOranı'] = np.round(grouped_df['ÇalışmaOranı'], 3) * 100
grouped_df['BoşGünler'] = grouped_df['FreeDay'] + grouped_df['OffDay']
#grouped_df['Gün'] = df.groupby(df['Tarih'].dt.to_period("M"))['Tarih'].count().reset_index()
grouped_df= grouped_df[['Tarih','Yapılan', 'Hedef', 'ÇalışmaOranı','FreeDay','OffDay','BoşGünler']]
grouped_df
```

```python
#Düzenle
df.loc[ (df['Yapılan'] == 0) & (df['Hedef']>0) , 'Yapılmayan'] = 1
df.loc[ (df['Yapılan'] > 0) , 'Yapılmayan' ] = 0
df.loc[ (df['FreeDay'] > 0) | (df['OffDay'] > 0), 'Yapılmayan'] = 0

df.loc[ (df['Yapılan'] > 0) & (df['Hedef']>0) , 'Yapıldı'] = 1
df.loc[ (df['Yapılan'] == 0) & (df['Hedef']>0) , 'Yapıldı'] = 0
df.loc[ (df['FreeDay'] > 0) | (df['OffDay'] > 0), 'Yapıldı'] = 0

df
```

```python
df['Yapılmayan'] = pd.to_numeric(df['Yapılmayan'], errors='coerce').astype('Int64')
df['Yapıldı'] = pd.to_numeric(df['Yapıldı'], errors='coerce').astype('Int64')
```
